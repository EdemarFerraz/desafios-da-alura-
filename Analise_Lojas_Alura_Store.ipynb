```python
import pandas as pd
import matplotlib.pyplot as plt

l1 = pd.read_csv('loja_1.csv')
l2 = pd.read_csv('loja_2.csv')
l3 = pd.read_csv('loja_3.csv')
l4 = pd.read_csv('loja_4.csv')

lojas_list = [l1, l2, l3, l4]
nomes = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

faturamentos = [df['Preço'].sum() for df in lojas_list]
fretes = [df['Frete'].mean() for df in lojas_list]
notas = [df['Avaliação da compra'].mean() for df in lojas_list]

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.bar(nomes, faturamentos, color=['#4CAF50', '#2196F3', '#FF9800', '#F44336'])
plt.title('Faturamento Total por Loja')
plt.ylabel('R$')

plt.subplot(1, 2, 2)
plt.scatter(fretes, notas, s=200, c=['#4CAF50', '#2196F3', '#FF9800', '#F44336'])
for i, txt in enumerate(nomes):
    plt.annotate(txt, (fretes[i], notas[i]), xytext=(5,5), textcoords='offset points')
plt.title('Custo de Frete vs Avaliação')
plt.xlabel('Frete Médio (R$)')
plt.ylabel('Nota Média')

plt.tight_layout()
plt.show()
```

1. Objetivo
O objetivo desta análise foi avaliar o faturamento e a eficiência operacional das quatro unidades da Alura Store para identificar qual delas deve ser vendida para o investimento em novos negócios.

2. Análise dos Dados
Ao observar os gráficos gerados, chegamos às seguintes conclusões:

Faturamento Total: A Loja 1 é a que possui o maior faturamento absoluto (ultrapassando R$ 1,5 milhão), seguida de perto pelas Lojas 2 e 3. A Loja 4 é a que possui o menor faturamento do grupo.

Eficiência Logística (Frete): A Loja 1 tem o frete médio mais caro de todos (quase R$ 35,00), o que é um ponto muito negativo para o cliente.

Satisfação (Avaliação): Apesar de faturar muito, a Loja 1 tem a pior nota média entre todas as unidades (abaixo de 3.98). Já a Loja 3 é a campeã de satisfação, com a maior nota.

Categorias e Produtos: Notamos que a categoria de Eletrodomésticos (como geladeiras e lavadoras) gera o maior faturamento devido ao alto valor unitário, enquanto Brinquedos e Livros têm grande volume, mas faturamento menor.

3. Conclusão e Recomendação
Minha recomendação para o Senhor João é a venda da Loja 1.

Justificativa:
Embora a Loja 1 tenha o maior faturamento, ela apresenta os sinais mais claros de desgaste operacional:

Pior Avaliação: É a loja que os clientes menos gostam, o que indica que o faturamento alto pode não se manter no futuro por falta de fidelidade.

Frete Abusivo: O custo de entrega é o mais alto do grupo, o que diminui a competitividade da unidade.

Oportunidade de Caixa: Como ela fatura muito, o valor de venda dessa unidade será maior, garantindo mais dinheiro para o Senhor João abrir o novo negócio, enquanto ele mantém as lojas 2, 3 e 4, que são mais eficientes e queridas pelos clientes.
